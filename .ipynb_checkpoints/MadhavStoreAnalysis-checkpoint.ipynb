```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
order=pd.read_csv('Orders.csv')
```

```python
order.head()
```

```python
pd.isnull(order)
```

```python
order.isnull().sum()
```

```python
detail=pd.read_csv('details.csv')
```

```python
detail.head()
```

```python
detail.shape
```

```python
order.shape
```

```python
df_inner=pd.merge(order,detail,how='inner',on='Order ID')
df_left=pd.merge(order,detail,how='left',on='Order ID')
df_right=pd.merge(order,detail,how='right',on='Order ID')
df_outer=pd.merge(order,detail,how='outer',on='Order ID')
```

```python
df_inner.shape
```

```python
df_left.shape
```

```python
df_right.shape
```

```python
df_outer.shape
```

```python
df_inner.isnull().sum()
```

```python
df_inner.head()
```

```python
# copy df in another 
df=df_inner.copy()
```

```python
df['Date']=pd.to_datetime(df['Order Date'],dayfirst=True)
```

```python
df['MonthName']=df['Date'].dt.month_name()
df['day'] = df['Date'].dt.day
df['year']=df['Date'].dt.year
df['daysname']=df['Date'].dt.day_name()
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop('Order Date',axis=1,inplace=True)
```

```python
df.head()
```

# EDA

# state wise analysis

```python
# state wise analysis
df_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=df_state,x='State',y='Amount',palette='Set2')
```

```python
df_state=df.groupby(['State'],as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=df_state,x='State',y='Quantity',palette='Set2')
```

```python
ax=sns.countplot(data=df,x='State')
sns.set(rc={'figure.figsize':(20,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df_state=df.groupby(['State'],as_index=False)['Quantity'].sum().sort_values(by='Quantity',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
ax=sns.barplot(data=df_state,x='State',y='Quantity',palette='Set2')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# state wise profit

df_state=df.groupby(['State'],as_index=False)['Profit'].sum().sort_values(by='Profit',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=df_state,x='State',y='Profit',palette='Set2')
```

```python
df_state_category=df.groupby(['State','Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=df_state_category,x='State',y='Amount',hue='Category',palette='Set2')
```

# Trend Analysis

```python
# yearly trend analysis

yearly_sales=df.groupby('year')['Amount'].sum().reset_index()

plt.figure(figsize=(20,6))
sns.lineplot(x='year',y='Amount',data=yearly_sales,marker='o')
plt.title('Yearly sales trend')
plt.xlabel('Year')
plt.ylabel('Total Sales (Amount)')
plt.show()
```

```python
# monthly sales trend

monthly_sales=df.groupby('MonthName')['Amount'].sum().reset_index()
plt.figure(figsize=(20,6))
sns.lineplot(x='MonthName',y='Amount',data=monthly_sales,marker='o')
plt.title('Monthly sales trend')
plt.xlabel('Month')
plt.ylabel('Total Sales (Amount)')
plt.show()
```

```python
# Days wise trends

days_trend=df.groupby('daysname')['Amount'].sum().reset_index()
plt.figure(figsize=(20,6))
sns.lineplot(x='daysname',y='Amount',data=days_trend,marker='o')
plt.title('days sales trend')
plt.xlabel('daysname')
plt.ylabel('Total Sales (Amount)')
plt.show()
```

```python
# daily sales by date and amount

daily_sales=df.groupby('Date')['Amount'].sum().reset_index()

plt.figure(figsize=(20,6))
sns.lineplot(x='Date',y='Amount',data=daily_sales,marker='o')
plt.title('Date sales trend')
plt.xlabel('Date')
plt.ylabel('Total Sales (Amount)')
plt.show()
```
